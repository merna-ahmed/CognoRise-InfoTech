```python

plt.bar('Region', 'Estimated Labour Participation Rate (%)' , data=df)
plt.title='Region by Estimated Labour Participation Rate (%)'
plt.xlabel='Region'
plt.ylabel='Estimated Labour Participation Rate (%)'
plt.xticks(rotation=90)
plt.show()

```

```python

plt.bar('Area', 'Estimated Employed', data=df , color='green')
plt.title='Region by Estimated Employer'
plt.xlabel='Area'
plt.ylabel='Estimated Employed'
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x='Region', y='Estimated Employed', data=df)
plt.title ='Boxplot of Estimated Employment by Region'
plt.xticks(rotation=90)
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Month']=df['Date'].dt.month
df.inplace=True
df
```

```python

plt.plot('Estimated Unemployment Rate (%)','Year' , data=df )
plt.title='Daily unemployment rate  over years'
plt.xlabel='Estimated Unemployment Rate (%)'
plt.ylabel='Year'
plt.grid(True)

plt.show()
```
